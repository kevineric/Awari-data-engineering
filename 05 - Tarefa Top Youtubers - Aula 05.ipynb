```python
import pandas as pd
import numpy as np
```

# COD PRONTO DAQUI PRA BAIXO

```python
df = pd.read_csv('C:/Users/kevin/topyoutubers.csv', delimiter="\t")
```

```python
df
```

```python
df['ident'] = df['Artist,Total Views,100M,Avg'].str.contains('"')
df
```

```python
df_true = df.query('ident == True')
df_true
```

```python
df_true['Artist'] = df_true['Artist,Total Views,100M,Avg'].str.split(',"').str.get(0)
df_true
```

```python
df_true['Total_Views'] = df_true['Artist,Total Views,100M,Avg'].str.split('"', expand=True).get(1)
df_true['Resto'] = df_true['Artist,Total Views,100M,Avg'].str.split('"', expand=True).get(2)
df_true[['Null', '100M', 'Avg']] = df_true['Resto'].str.split(',', expand=True)

df_true
```

```python
df_true2 = df_true[['Artist','Total_Views','100M', 'Avg']]
df_true2['Total_Views'] = df_true2['Total_Views'].str.replace('.',' ')
df_true2['Total_Views'] = df_true2['Total_Views'].str.replace(',','')
df_true2['Total_Views'] = df_true2['Total_Views'].str.replace(' ','.')

df_true2
```

```python
df_true2['Total_Views'] = df_true2['Total_Views'].astype(float)
df_true2['100M'] = df_true2['100M'].apply(lambda x: x if x != '' else '0')
df_true2['100M'] = df_true2['100M'].astype(int)
df_true2['Avg'] = df_true2['Avg'].astype(float)
df_true2
```

```python
df_true2.dtypes
```

```python
df_true[df_true['Artist,Total Views,100M,Avg'].str.contains('Earth')==True]
```

```python
df_false = df.query('ident == False')
df_false
```

```python
df_false[['Artist', 'Total_Views', '100M', 'Avg']] = df_false['Artist,Total Views,100M,Avg'].str.split(',', expand=True)
df_false2 = df_false[['Artist', 'Total_Views', '100M', 'Avg']]
df_false2['Total_Views'] = df_false2['Total_Views'].astype(float)
df_false2['100M'] = df_false2['100M'].apply(lambda x: x if x != '' else '0')
df_false2['100M'] = df_false2['100M'].astype(float)
df_false2['Avg'] = df_false2['Avg'].astype(float)

df_false2
```

```python
df_false2.dtypes
```

```python
df_consolidado = pd.concat([df_true2,df_false2])
df_consolidado
```

### Tratando vazios com zeros nas colunas float

```python
df_consolidado['Total_Views'] = df_consolidado['Total_Views'].fillna(0)
df_consolidado['Avg'] = df_consolidado['Avg'].fillna(0)
df_consolidado
```

### Multiplicando coluna por 100.000.000 e coluna Avg com 2 casas decimais

```python
df_consolidado['100M'] = df_consolidado['100M'] * 100000000
df_consolidado['Avg'] = df_consolidado['Avg'].round(2)
pd.options.display.float_format = '{:.2f}'.format


df_consolidado
```

### Top 100 da lista ordenado por nome de forma decrescente

```python
df_consolidado = df_consolidado.sort_values(by='Artist',ascending=False)
df_consolidado.head(100)
```

### Exportando arquivo tratado para CSV

```python
df_consolidado = df_consolidado.sort_values(by='Total_Views',ascending=False)
df_consolidado.to_csv('C:/Users/kevin/topyoutubers_trat.csv', index=False)
```
